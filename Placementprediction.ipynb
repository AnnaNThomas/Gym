```python
X = data.drop('status',axis=1)
y= data['status']
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn import svm
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
# Gemini Version
```

```python
 X_train = X_train.apply(pd.to_numeric, errors='coerce')
```

```python
X_train = X_train.fillna(0)
```

```python
# github
lr = LogisticRegression()
lr.fit(X_train,y_train)

svm = svm.SVC()
svm.fit(X_train,y_train)

knn=KNeighborsClassifier()
knn.fit(X_train,y_train)

dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)

rf=RandomForestClassifier()
rf.fit(X_train,y_train)

gb=GradientBoostingClassifier()
gb.fit(X_train,y_train)
```

```python
# Gemini
```

```python
X_test = X_test.apply(pd.to_numeric, errors='coerce')
```

```python
X_test = X_test.fillna(0)
```

```python
#github
y_pred1 = lr.predict(X_test)
y_pred2 = svm.predict(X_test)
y_pred3 = knn.predict(X_test)
y_pred4 = dt.predict(X_test)
y_pred5 = rf.predict(X_test)
y_pred6 = gb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
score1=accuracy_score(y_test,y_pred1)
score2=accuracy_score(y_test,y_pred2)
score3=accuracy_score(y_test,y_pred3)
score4=accuracy_score(y_test,y_pred4)
score5=accuracy_score(y_test,y_pred5)
score6=accuracy_score(y_test,y_pred6)
```

```python
print(score1,score2,score3,score4,score5,score6)
```

```python
final_data = pd.DataFrame({'Models':['LR','SVC','KNN','DT','RF','GB'],
            'ACC':[score1*100,
                  score2*100,
                  score3*100,
                  score4*100,
                  score5*100,score6*100]})
```

```python
#chatgpt
from pyngrok import ngrok

# Streamlit app code
streamlit_code = """
import streamlit as st
import joblib
import pandas as pd

def make_prediction(data):
    model = joblib.load('model_campus_placement.pkl')
    result = model.predict(data)
    result_proba = model.predict_proba(data)
    return result[0], result_proba[0][1]

def main():
    st.title("Campus Placement Prediction System")

    gender = st.selectbox("Gender", ["Male", "Female"])
    ssc_p = st.number_input("Secondary Education percentage - 10th Grade", min_value=0.0, max_value=100.0)
    ssc_b = st.selectbox("Board of Education (10th Grade)", ["Central", "Others"])
    hsc_p = st.number_input("Higher Secondary Education percentage - 12th Grade", min_value=0.0, max_value=100.0)
    hsc_b = st.selectbox("Board of Education (12th Grade)", ["Central", "Others"])
    hsc_s = st.selectbox("Specialization in Higher Secondary Education", ["Science", "Commerce", "Arts"])
    degree_p = st.number_input("Degree Percentage", min_value=0.0, max_value=100.0)
    degree_t = st.selectbox("Under Graduation (Degree type) - Field of degree education", ["Sci&Tech", "Comm&Mgmt", "Others"])
    workex = st.selectbox("Work Experience", ["Yes", "No"])
    etest_p = st.number_input("Enter test percentage", min_value=0.0, max_value=100.0)
    specialisation = st.selectbox("Branch specialization", ["Mkt&HR", "Mky&Fin"])
    mba_p = st.number_input("MBA percentage", min_value=0.0, max_value=100.0)

    gender_mapping = {"Male": 1, "Female": 0}
    board_mapping = {"Central": 1, "Others": 0}
    specialization_mapping = {"Science": 2, "Commerce": 1, "Arts": 0}
    degree_mapping = {"Sci&Tech": 2, "Comm&Mgmt": 1, "Others": 0}
    yes_no_mapping = {"Yes": 1, "No": 0}
    mkt_hr_mapping = {"Mkt&HR": 1, "Mky&Fin": 0}

    input_data = pd.DataFrame({
        'gender': [gender_mapping[gender]],
        'ssc_p': [ssc_p],
        'ssc_b': [board_mapping[ssc_b]],
        'hsc_p': [hsc_p],
        'hsc_b': [board_mapping[hsc_b]],
        'hsc_s': [specialization_mapping[hsc_s]],
        'degree_p': [degree_p],
        'degree_t': [degree_mapping[degree_t]],
        'workex': [yes_no_mapping[workex]],
        'etest_p': [etest_p],
        'specialisation': [mkt_hr_mapping[specialisation]],
        'mba_p': [mba_p],
    })

    if st.button("Predict"):
        result, result_proba = make_prediction(input_data)
        if result == 0:
            st.write("Can't Be Placed")
        else:
            st.write(f"Student Will be Placed With Probability of {round(result_proba * 100, 2)}%")

if __name__ == "__main__":
    main()
"""

with open('app.py', 'w') as file:
    file.write(streamlit_code)
```
